```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
plt.style.use('seaborn-whitegrid')
```

```python
dfnym=pd.read_csv('tablanym.csv')
dfny=pd.read_csv('tablany.csv')
dfnym.rename(columns={'Win%[11]':'Winsnym'},inplace=True)
dfny.rename(columns={'Win%[11]':'Winsny'},inplace=True)
dfny.sort(['Yankees season'])
dfnym.sort(['Mets season'])
nym=dfnym[dfnym['Mets season']>=1970]
len(nym)
ny=dfny[dfny['Yankees season']>=1970]
len(ny)
ny=ny.sort(['Yankees season'])
ny
nym=nym.sort(['Mets season'])
nym
nym=nym.set_index([nym['Mets season']])
nym['Winsnym']
ny=ny.set_index([ny['Yankees season']])
ny['Winsny']
```

```python
plt.plot(ny['Winsny'])
plt.plot(nym['Winsnym'],color='red')
plt.xticks(range(1970,2016,5),rotation='45')
plt.legend(['New York Yankees','New York Mets'],loc='lower center')
plt.gca().axis([1970, 2016, 0, 0.9])
plt.ylabel('Winning Percentage')
plt.xlabel('Years')
plt.title('Comparasion between the New York Mets and New York Yankees\n winning record\n 1970-2016 ')
```
